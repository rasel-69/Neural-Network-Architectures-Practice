```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
```

```python
class SimpleResidualBlock(nn.Module):
    def __init__(self, features):
        super(SimpleResidualBlock, self).__init__()

        self.fc1 = nn.Linear(features, features)
        self.fc2 = nn.Linear(features, features)
        self.relu = nn.ReLU()

    def forward(self, x):
        identity = x                 # üîÅ Skip connection

        out = self.relu(self.fc1(x))
        out = self.fc2(out)

        out = out + identity         # ‚ûï Add skip connection
        out = self.relu(out)

        return out
```

```python
class FiveLayerNN(nn.Module):
    def __init__(self, input_size, num_classes):
        super(FiveLayerNN, self).__init__()

        # Layer 1
        self.fc1=nn.Linear(input_size, 128)
        # Layer 2
        self.fc2=nn.Linear(128,128)
        
        # Layer 3 and 4 (Residual Block)
        self.residual_block=SimpleResidualBlock(128)

        # Layer 5 output
        self.fc_out=nn.Linear(128, num_classes)

    def forward(self, x):
        x=self.relu(self.fc1(x))  #layer 1
        x=self.relu(self.fc2(x))  #layer 2
        x=self.residual_block(x)  # residual block 3 and 4
        x=self.fc_out(x)

        return x 
```
